```
fruits=["Banana","Mango","Pear","Apple","Kiwi","Grape"]

count=3

while count>0:

  try:
    index=int(input("Please enter a favorite fruit: "))
    print("My favorite fruit is {}".format(fruits[index]))
    break

  except IndexError:
    count -=1
    print(f"There is no such index. You have {count} right/s to enter an index")

  except ValueError:
    count-=1
    print(f"Please enter an integer number. You have {count} right/s") 

  else:
    print("Congratulations!! You have entered correctly")
    break

  finally:
    print("My fruits are delicious!!!")   
```

```
names=["ali","veli"] # True - If all elements in an iterable are true
mood=["happy","sad",0] # False - If any element in an iterable is false
empty={}

print(all(names),all(mood),all(empty),sep="\n") # all sadece Truty ve Falsy verir
```

```
names=["ali","veli", False] # True - If at least one element of an iterable is true
mood=["happy","sad",0] # False - If all elements are false or if an iterable is empty
empty={}

print(any(names),any(mood),any(empty),sep="\n") # any sadece Truty ve Falsy verir
```

```
anagrams=["eat","tea","tan","ate","nat","bat"]

grouped_anagrams = {}

for i in anagrams:
   
   sorted_string = str(sorted(i))
  
   if sorted_string in grouped_anagrams:
     
      grouped_anagrams[sorted_string].append(i)

   else:
         
         grouped_anagrams[sorted_string] = [i]

print(list(grouped_anagrams.values()))    

  
```

```
def equal(*a):                  # en fazla tekrar eden sayıyı bulma
  numbers=list(a)
  result=numbers.count(max(numbers,key=numbers.count))

  if result>1:
    return result
  else:
    return 0  
```
