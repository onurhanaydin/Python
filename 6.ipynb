```python
#left finger on the keyboard

left={"q","w","e","r","t","a","s","d","f","g","z","x","c","v","b"}

#right finger on the keyboard

right={"y","u","o","p","h","j","k","l","i","n","m"}

word=input("Please enter a word: ")

inword=set(word)

a=set(inword|left) 
b=set(inword|right)

comf_word=(a.issubset(left)) or (b.issubset(right)) #if it uses only one hand keys, it would be called as a comfartable word

print(comf_word)
```
