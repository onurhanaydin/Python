```python
#=======================================================================
# Written by Onurhan AYDIN for Liminalytics, March 2022
# Renaming files from IPCC atlas
#=======================================================================
# Developer area
#=======================================================================
#
# load Python libraries 
import glob
import shutil
#=======================================================================
# User area: provide parameters
#=======================================================================
give_var = "Surface wind  ms_-1" # description/full name of the variable

#=======================================================================
# Developer area: describe the data
#=======================================================================
# Some lists and vocabularies describing the data
# Since "_hist" will be added in history section, "CMIP6_hist" #"CORDEX EUROPE_hist" were not used
datasets = ["CMIP6", "CORDEX Europe"] #"CMIP6_hist" #"CORDEX EUROPE_hist" 

voc_periods = {"Long Term (2081-2100)":"Long" , "Medium Term (2041-2060)":"Medium","Near Term (2021-2040)":"Near"}

seasons = {"December to February" , "March to May", "June to August", "September to November", "Annual"}

voc = {"Surface wind  ms_-1": "surf_wind", "CORDEX Europe":"CORDEX_Europe", "Long Term (2081-2100)":"2081-2100", 
       "Medium Term (2041-2060)":"2041-2060", "Near Term (2021-2040)":"2021-2040", "Annual":"annual",
       "December to February":"winter" , "March to May":"spring", "June to August":"summer", "September to November":"autumn", "CMIP6":"CMIP6"}

   
#=======================================================================
# structure of the data
#=======================================================================
newfoldername = "processed/" # The folder must be created as a subfolder

# VERY IMPORTANT!!
# pay attention to the structure and names of the folders:
# ls:    
#CMIP6                                     CORDEX Europe_hist
#CMIP6_hist                                database_IPCC_climatology_rename_wind.py
#CORDEX Europe                             processed

#=======================================================================
# actual renaming
#=======================================================================
for dataset in datasets:
    
    if dataset == "CMIP6":
        scenarios = {"SSP1-2.6", "SSP2-4.5", "SSP3-7.0", "SSP5-8.5"}
        periods = {"Long Term (2081-2100)" , "Medium Term (2041-2060)","Near Term (2021-2040)"}
        periods_hist = {"1850-1900" , "1961-1990","1981-2010", "1986-2005", "1995-2014"}        
    
    if dataset == "CORDEX Europe":
        scenarios = {"RCP2.6","RCP4.5","RCP8.5" }
        periods = {"Long Term (2081-2100)" , "Medium Term (2041-2060)","Near Term (2021-2040)"}
        periods_hist = {"1981-2010", "1986-2005", "1995-2014"}        
#=======================================================================        
    # future
    for scenario in scenarios: 
        for period in periods: 
            for season in seasons:
                foldername = "/Users/onurhanaydin/Desktop/Liminalytics/Datasets/surf_wind/" + dataset + "/" + scenario + "/" + voc_periods[period] + "/"
                filename = foldername + dataset + " - " + give_var + " - " + period + " " + scenario + "  - " + season
                loadfile = glob.glob(filename + '*.nc')
                #print("I look for ", filename)
                #print("I loaded", loadfile)
                filename_new = "/Users/onurhanaydin/Desktop/Liminalytics/Datasets/surf_wind/" + newfoldername + voc[dataset] + "_" + voc[give_var] + "_" + voc[period] + "_" + scenario + "_" + voc[season] + ".nc"
                #print("Newname is ", filename_new)
                
                src = str(loadfile[0])
                dest = filename_new
    
                shutil.copy(src, dest)
#=======================================================================
    # historical 
    for period in periods_hist: 
        for season in seasons:
            foldername = "/Users/onurhanaydin/Desktop/Liminalytics/Datasets/surf_wind/" + dataset + "_hist/" + period + "/"
            filename = foldername + dataset + " - " + give_var + " - " + period + "   - " + season
            loadfile = glob.glob(filename + '*.nc')
            #print("I look for ", filename)
            #print("I loaded", loadfile)
            filename_new = "/Users/onurhanaydin/Desktop/Liminalytics/Datasets/surf_wind/" + newfoldername + voc[dataset] + "_" + voc[give_var] + "_" + period +"_hist" + "_" + voc[season] + ".nc"
            #print("Newname is ", filename_new)
            
            src = str(loadfile[0])
            dest = filename_new
    
            shutil.copy(src, dest)            
#=======================================================================
```
